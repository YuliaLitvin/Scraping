```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://letual.com.ua/services/'
```

```python
headers = {
    "Accept": "*/*",
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
```

```python
req = requests.get(url, headers)
```

```python
req
```

```python
req.text
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
soup
```

```python
soup.find_all(class_='service__item')
```

```python
dct = {}
for item in soup.find_all(class_='service__item'):
    name = item.find(class_='service__name').text
    price = item.find(class_='service__price').text
    dct[name] = price
    
    
    print(name)
    print(price)
    print('\n')
```
